```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/content/gdrive/MyDrive/data sigma/acquisitions_update_2021.csv")
```

```python
ibm = df.loc[df['Parent Company']=='Ibm']
ibm = ibm.replace("-", np.nan)
ibm =ibm.replace("undisclosed",np.nan)
```

```python
plt.figure(figsize= (15,10))
ax=sns.countplot(x='Acquisition Year',data=ibm)
for p in ax.patches:
   ax.annotate('{:.0f}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.5))
```

```python
# try to add colour theme, all 3 pi charts same line (subplots)
countries = (129,23,10,13,3,2)
mylabels = ["USA", "Europe", "Asia","Canada","Australia", "Brazil"]
explode = (0.2, 0.2,0.2,0.4,0.4,0.45)
plt.pie(countries, labels = mylabels, radius=2.5, explode= explode, autopct='%1.2f%%', shadow= True)
plt.legend(title = "IBM Aquisitions in different countries", bbox_to_anchor=(2.8,1.5), loc="upper right")
plt.show()
```

```python
#europe - colour theme, subplots, increase figure size
europelabels=["UK", "Germany","France","Switzerland","Italy","Ireland","Norway",
"Netherlands", "Denmark", "Finland" , "Sweden","Spain"]
europe= (4,4,3,2,2,2,1,1,1,1,1,1)

plt.pie(europe, labels= europelabels, radius=1.5, autopct='%1.2f%%')
plt.title("IBM Acquisitions in Europe", loc= "center",pad = 60, fontsize='20',
          backgroundcolor='black',
          color='white')

centre_circle = plt.Circle((0,0),1.2,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)


plt.show()
```

```python
#asia- subplot, theme, increase figsize
asialabels=["Israel", "India", "Hong Kong", "Singapore"]
asia= (5,3,1,1)

plt.pie(asia, labels= asialabels, radius=1.5, autopct='%1.2f%%')
plt.title("IBM Acquisitions in Asia", loc= "center",pad = 60, fontsize='20',
          backgroundcolor='black',
          color='white')
centre_circle = plt.Circle((0,0),1.2,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.show()
```

```python
month = ibm['Acquisition Month'].value_counts()
month_ibm = month.to_frame().reset_index()
month_ibm = month_ibm.rename({'index': 'months', 'Acquisition Month': 'Total'}, axis=1)
month_ibm = month_ibm.set_index('months')
month_ibm.index = pd.CategoricalIndex(month_ibm.index, categories=['Jan', 'Feb', 'Mar', 'Apr','May',
                                          'Jun', 'Jul', 'Aug','Sep', 'Oct', 'Nov', 'Dec'], ordered=True)
month_ibm = month_ibm.sort_index()
print(month_ibm)

plt.figure(figsize=(15,10))
plt.bar(x=month_ibm.index, height='Total', data=month_ibm, color='crimson')
plt.title('Months With Highest Number of Acquisitions')
plt.xlabel('Months')
plt.ylabel('Total Acquisitions');
```

```python
business = ibm['Business'].value_counts()
business_ibm = business.to_frame().reset_index()
business_ibm = business_ibm.rename({'index': 'Business', 'Business': 'Total'}, axis=1)
print(business_ibm)
plt.figure(figsize=(10,15))
plt.title('Businesses acquired by IBM')
sns.barplot(x='Total', y='Business', data=business_ibm[:31]);
```
